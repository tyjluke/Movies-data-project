```python
# what genre of movies uses the largest budget
# does a larger budget get more votes?

# corrplations
#First I imported the packages for use in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



#read in the data
df = pd.read_csv('file:///Users/tylukasiewicz/Downloads/movies.csv')
```

```python
#looking at the data

df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#quick look at what are data types are

df.dtypes
```

```python
# here I am just making the year column into four integer number to clean up the data

df['Year'] = df['released'].astype(str).str[:4]
df
```

```python
# More cleaning of data by dropping any dupicates

df.drop_duplicates()
```

```python
# Order our Data a little bit to see movies with highest budget

df.sort_values(by=['budget'], inplace=False, ascending=False)
```

```python
# ok now I will create a scatter plot of movie genre and movie budget to see if there is a quick cooralation 

plt.scatter(x=df['budget'], y=df['genre'], alpha=0.5)
plt.title('Budget vs Genre')
plt.xlabel('budget for Film')
plt.ylabel('genre of Film')
plt.show()
```

```python
#lets look at the cooralation
```

```python
# Here I will take a look at budget and votes by creating a regplot 

sns.regplot(x="budget", y="votes", data=df)
```

```python
# I notice there is a positive cooralation between votes, and bueget.
```

```python
df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)


strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
# Lastly we will change all fields to numerical values and create a heatmat to show the final cooralations

df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df_numerized.corr(method='pearson')
```

```python
# creating the heat map here 

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Finnally we are able to sumerize our findings.

# we found a positive cooralation between action genre movies and budget. where the budget is normally higher then other films.
# We also found that having a larger budget will corelate to a greater inverval of votes. 
```
